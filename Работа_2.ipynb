# **俄罗斯数据集(Русский набор данных)**

```python
#读取wildberries数据集(Прочитайте набор данных wildberries)
csv_wildberries = "/content/wildberries宠物用品数据.csv"
csv_data_wildberries = pd.read_csv(csv_wildberries, low_memory = False)
csv_df_wildberries = pd.DataFrame(csv_data_wildberries)
csv_df_wildberries
```

数据清洗

```python
# 存储处理后的数据（Сохраняем обработанные данные）
wildberries_data_1 = []

# 打开名为csv_wildberries的文件，以只读模式打开，使用utf-8编码（Откройте файл csv_wildberry, откройте его в режиме только для чтения, используя кодировку utf-8.）
with open(csv_wildberries, 'r', encoding = 'utf-8') as file:
    # 创建一个CSV字典阅读器对象（Создайте объект чтения словаря CSV.）
    reader = csv.DictReader(file)
    # 遍历读取每一行数据，将其作为字典添加到列表wildberries_data_1中（Просматривайте и читайте каждую строку данных и добавляйте ее как словарь в список wildberry_data_1.）
    for row in reader:
        wildberries_data_1.append(row)


# 删除重复值（Удалить повторяющиеся значения）
csv_df_wildberries.drop_duplicates(inplace=True)

# 将数据集中的大写字母转换为小写字母（Преобразование прописных букв в наборе данных в строчные буквы）
csv_df_wildberries = csv_df_wildberries.apply(lambda x: x.astype(str).str.lower())
```

列“brand”（Список “brand”）

```python
# 删除列“brand"为空值的行(Удалить строки, в которых столбец «бренд» имеет пустые значения.)
csv_df_wildberries = csv_df_wildberries.dropna(subset=['brand'])
```

列“price”（Список “price”）

```python
# 删除列“price"为空值的行(Удалить строки, в которых столбец «цена» имеет пустые значения.)
csv_df_wildberries = csv_df_wildberries.dropna(subset=['price'])

# 将列“price”转换为整数型（Преобразовать столбец «цена» в целочисленный тип）
csv_df_wildberries['price'] = csv_df_wildberries['price'].astype(int)
```

列“count”（Список “count”）

```python
# 将列“count”转换为浮点数型，将无法转换的非数值设为 NaN（Преобразуйте столбец «Комментарий» в число с плавающей запятой, установите нечисловые значения, которые невозможно преобразовать в NaN）
csv_df_wildberries['count'] = pd.to_numeric(csv_df_wildberries['count'], errors='coerce')
# 删除列“count”为空值的行（Удалить строки с нулевым значением в столбце «Комментарий»）
csv_df_wildberries = csv_df_wildberries.dropna(subset=['count'])
# 将列“count”转换为整数型（Преобразовать столбец «Комментарий» в целочисленный тип）
csv_df_wildberries['count'] = csv_df_wildberries['count'].astype(int)
```

列“rating”（Список “rating”）

```python
# 将列“rating"中的“,”替换为“.”（Замените «,» в столбце «материал» на «.».）
csv_df_wildberries['rating'] = csv_df_wildberries['rating'].str.replace(',', '.', regex=False)

# 将列“rating”转换为浮点数型，将无法转换的非数值设为 NaN（Преобразуйте столбец «Комментарий» в число с плавающей запятой, установите нечисловые значения, которые невозможно преобразовать в NaN）
csv_df_wildberries['rating'] = pd.to_numeric(csv_df_wildberries['rating'], errors='coerce')
```

列“origin”（Список “origin”）

```python
# 从info_dict中提取信息（Извлеките информацию из info_dict.）
def extract_info_r(info_str, key):
    info_dict = literal_eval(info_str)
    return info_dict.get(key, None)

# 为数据集添加列“origin" (Добавить столбец «origin» в набор данных)
csv_df_wildberries['origin'] = csv_df_wildberries['info_dict'].apply(lambda x: extract_info_r(x, 'страна производства'))
```

列“Animal species”（Список “Animal species”）

```python
# 为数据集添加列“Animal species"(Добавить столбец «Виды животных» в набор данных)
csv_df_wildberries['Animal_species'] = csv_df_wildberries['info_dict'].apply(lambda x: extract_info_r(x, 'вид животного'))

# 去除"для"
csv_df_wildberries["Animal_species"] = csv_df_wildberries["Animal_species"].str.replace("для", "", regex=False)
# 去除空格
csv_df_wildberries["Animal_species"] = csv_df_wildberries["Animal_species"].str.replace(" ", "", regex=False)
```

删除特征列和调整特征列顺序及保存修改后的数据集（Удалить столбцы признаков, изменить порядок столбцов признаков и сохранить измененный набор данных.）

```python
# 删除“info_dict"列(Удалить столбец "info_dict")
csv_df_wildberries.drop(columns = [ 'info_dict'], inplace = True)

# 调整列的顺序(Отрегулируйте порядок столбцов)
new_column_order = ['title', 'brand', 'origin', 'Animal_species', 'count', 'rating', 'price']
wildberries_data_1 = csv_df_wildberries[new_column_order]

# 处理后的数据存储到新的CSV文件中(Обработанные данные сохраняются в новом CSV-файле.)
output_csv_wildberries = '/content/wildberries宠物用品数据（处理后1.0）.csv'
fieldnames = ['title', 'brand', 'origin', 'Animal_species', 'count', 'rating', 'price']

with open(output_csv_wildberries, 'w', encoding = 'utf-8', newline= '') as file:
    writer_wildberries = csv.DictWriter(file, fieldnames = fieldnames)

    # 写入表头(Написать заголовок)
    writer_wildberries.writeheader()

    # 写入处理后的数据(Запись обработанных данных)
    writer_wildberries.writerows(wildberries_data_1.to_dict(orient = 'records'))
```
